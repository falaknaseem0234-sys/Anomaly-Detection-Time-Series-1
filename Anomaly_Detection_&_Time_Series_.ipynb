Question 1:  What is Anomaly Detection? Explain its types (point, contextual, and
collective anomalies) with examples.
Ans 1Anomaly Detection

Anomaly Detection is a data analysis technique used to identify data points, patterns, or observations that deviate significantly from the normal or expected behavior of a dataset. Such unusual observations are called anomalies or outliers. Anomaly detection is crucial in many real-world applications because anomalies often indicate important events such as fraud, system failures, cyber-attacks, or medical abnormalities.

Types of Anomalies

Anomalies are generally classified into three main types:

1. Point Anomalies
Definition

A point anomaly is a single data instance that is significantly different from the rest of the data.

Explanation

Here, an individual data point is considered abnormal when compared with the overall data distribution, regardless of context.

Example

Credit card fraud:
A transaction of ‚Çπ5,00,000 made using a credit card when the customer usually spends around ‚Çπ2,000‚Äì‚Çπ5,000.

Temperature monitoring:
A temperature reading of 90¬∞C in a room where the normal range is 20‚Äì30¬∞C.

Network traffic:
A sudden spike in data packets from a single IP address.

Key Characteristic

Anomaly can be detected by observing one data point alone.

2. Contextual Anomalies (Conditional Anomalies)
Definition

A contextual anomaly is a data point that is abnormal only within a specific context, such as time, location, or season.

Explanation

The same data value may be normal in one context but anomalous in another. Contextual anomalies require contextual attributes (e.g., time, date, location) and behavioral attributes (e.g., value being measured).

Example

Weather data:
A temperature of 35¬∞C is normal in summer but anomalous in winter.

E-commerce sales:
High sales at midnight may be anomalous for a physical store but normal during festival sales.

Healthcare monitoring:
A heart rate of 110 bpm is normal during exercise but anomalous during sleep.

Key Characteristic

Anomaly depends on context, not just the value.

3. Collective Anomalies
Definition

A collective anomaly occurs when a group of data points together is anomalous, even if individual points appear normal.

Explanation

The anomaly is identified only when considering the sequence, pattern, or structure of data points as a whole.

Example

Time-series data:
A series of small but continuous network requests that together indicate a DDoS attack.

ECG signals:
A sequence of normal heartbeats forming an abnormal rhythm.

Stock market:
A prolonged downward trend indicating market manipulation, even if daily changes are small.

Key Characteristic

Individual points may be normal, but their combined behavior is abnormal.

Conclusion

Anomaly detection helps uncover unusual patterns that often signal critical issues.

Point anomalies focus on individual abnormal data points.

Contextual anomalies depend on the situation or environment.

Collective anomalies involve abnormal patterns formed by groups of data points.

Understanding these types enables the selection of appropriate anomaly detection techniques for real-world applications such as fraud detection, healthcare monitoring, cybersecurity, and industrial systems.

Question 2: Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of
their approach and suitable use cases.
Ans 2nomaly detection techniques differ in their underlying approach, assumptions, and suitable applications. Isolation Forest, DBSCAN, and Local Outlier Factor (LOF) are three widely used methods for detecting anomalies in data. The comparison below explains their working principles and appropriate use cases.

1. Isolation Forest
Approach

Isolation Forest is a tree-based, ensemble method specifically designed for anomaly detection.

It works on the idea that anomalies are easier to isolate than normal points.

The algorithm randomly selects a feature and a split value to build isolation trees.

Anomalies require fewer splits to be isolated and therefore have shorter path lengths in the trees.

Key Characteristics

Does not rely on distance or density measures.

Scales well to large and high-dimensional datasets.

Efficient and computationally fast.

Suitable Use Cases

Fraud detection in banking and financial transactions.

Network intrusion detection with large volumes of data.

High-dimensional data (e.g., log files, sensor readings).

Situations where labeled data is unavailable.

2. DBSCAN (Density-Based Spatial Clustering of Applications with Noise)
Approach

DBSCAN is a density-based clustering algorithm.

It groups together points that are densely packed and labels points in low-density regions as noise (anomalies).

Uses two parameters:

Œµ (epsilon): neighborhood radius

MinPts: minimum number of points to form a dense region

Key Characteristics

Detects clusters of arbitrary shapes.

Explicitly identifies outliers as noise.

Does not require the number of clusters in advance.

Suitable Use Cases

Spatial data analysis (e.g., GPS coordinates, geographic data).

Image segmentation and pattern recognition.

Anomaly detection where normal data forms dense clusters.

Small to medium-sized datasets with clear density separation.

3. Local Outlier Factor (LOF)
Approach

LOF is a density-based, local anomaly detection method.

It compares the local density of a data point with the densities of its neighbors.

A point is considered an anomaly if it has significantly lower density than its neighbors.

Key Characteristics

Focuses on local structure, not global distribution.

Effective in datasets with varying densities.

Produces an outlier score rather than a binary decision.

Suitable Use Cases

Medical diagnosis (e.g., unusual patient measurements).

Intrusion detection in systems with evolving behavior.

Datasets where anomalies are relative to nearby points, not the entire dataset.

Medium-sized datasets requiring fine-grained detection.

Scalability	Very high	Moderate	Moderate
Detects Local Anomalies	Moderate	Limited	‚úî Excellent
Parameter Sensitivity	Low	High (Œµ, MinPts)	Medium (k neighbors)
Conclusion

Isolation Forest is best for large-scale, high-dimensional anomaly detection where efficiency is critical.

DBSCAN is suitable when anomalies lie in low-density regions and the data has clear cluster structure.

Local Outlier Factor excels at detecting context-dependent, local anomalies in datasets with varying densities.

Each method has unique strengths, and the choice depends on data size, dimensionality, and anomaly characteristics.

Question 3: What are the key components of a Time Series? Explain each with one
example.
Ans 3 Key Components of a Time Series

A time series is a sequence of observations recorded at regular time intervals (hourly, daily, monthly, yearly, etc.). Time series analysis helps in understanding underlying patterns in data and in forecasting future values.
A time series is generally decomposed into four key components.

1. Trend (T)
Explanation

The trend represents the long-term movement or direction of the data over time. It shows whether the series is generally increasing, decreasing, or remaining stable.

Example

Annual sales of electric vehicles increasing steadily over several years due to technological advancement and environmental awareness.

Example trend: Sales rise from 10,000 units in 2015 to 80,000 units in 2024.

2. Seasonal Component (S)
Explanation

The seasonal component refers to regular and predictable patterns that repeat at fixed intervals within a specific period (daily, weekly, monthly, or yearly).

Example

Retail sales increase every year during festivals like Diwali or Christmas.

Ice cream sales peak every summer and drop in winter.

3. Cyclical Component (C)
Explanation

The cyclical component represents long-term fluctuations caused by economic or business cycles. Unlike seasonality, cycles do not have a fixed or known period.

Example

GDP growth rates rising during economic booms and falling during recessions.

Employment rates increasing during economic expansion and decreasing during downturns.

4. Irregular (Random) Component (I)
Explanation

The irregular component consists of unpredictable, random variations that cannot be explained by trend, seasonality, or cycles. These are often caused by unexpected events.

Example

A sudden drop in airline bookings due to a natural disaster or pandemic.

Stock price fluctuations caused by unexpected political announcements.

Question 4: Define Stationary in time series. How can you test and transform a
non-stationary series into a stationary one?
Ans 4 Stationarity in Time Series

A time series is said to be stationary if its statistical properties remain constant over time. In other words, the series does not change its behavior as time progresses, which makes it easier to model and forecast.

Definition of a Stationary Time Series

A time series is stationary if it satisfies the following conditions:

Constant Mean ‚Äì The average value of the series does not change over time.

Constant Variance ‚Äì The variability of the series remains the same over time.

Constant Autocovariance ‚Äì The correlation between values depends only on the lag, not on time.

Example of stationary series:
Daily fluctuations of a stable sensor signal around a fixed mean.

Example of non-stationary series:
Stock prices with upward trend and changing variance.

Why Stationarity is Important

Many time series models (ARIMA, SARIMA) assume stationarity.

Non-stationary data can lead to misleading forecasts and spurious relationships.

Stationary series are easier to analyze and interpret.

Testing Stationarity
1. Visual Inspection

Time plot: Look for trends, seasonality, or changing variance.

Rolling statistics: Plot rolling mean and rolling variance; they should remain constant for stationary data.

2. Statistical Tests
(a) Augmented Dickey‚ÄìFuller (ADF) Test

Null Hypothesis (H‚ÇÄ): The series is non-stationary (has a unit root).

Alternative Hypothesis (H‚ÇÅ): The series is stationary.

If p-value < 0.05, reject H‚ÇÄ ‚Üí series is stationary.

(b) KPSS Test

Null Hypothesis (H‚ÇÄ): The series is stationary.

Alternative Hypothesis (H‚ÇÅ): The series is non-stationary.

If p-value < 0.05, reject H‚ÇÄ ‚Üí series is non-stationary.

Using both ADF and KPSS gives more reliable conclusions.

Question 6: Load a time series dataset (e.g., AirPassengers), plot the original series,
and decompose it into trend, seasonality, and residual components
(Include your Python code and output in the code box below.)

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose


data = pd.read_csv(
    "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv"
)

# Convert Month column to datetime
data['Month'] = pd.to_datetime(data['Month'])
data.set_index('Month', inplace=True)

# Plot original time series
plt.figure()
plt.plot(data['Passengers'])
plt.title("AirPassengers Time Series")
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.show()

# Decompose the time series
decomposition = seasonal_decompose(
    data['Passengers'],
    model='multiplicative',
    period=12
)

# Plot decomposition
decomposition.plot()
plt.show()
```

Question 7: Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) to
detect anomalies. Visualize the anomalies on a 2D scatter plot.
(Include your Python code and output in the code box below.)

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# Load sample NYC Taxi Fare dataset
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/taxis.csv"
data = pd.read_csv(url)

# Select numerical features
df = data[['distance', 'fare']].dropna()

# Apply Isolation Forest
iso_forest = IsolationForest(
    n_estimators=100,
    contamination=0.05,
    random_state=42
)

df['anomaly'] = iso_forest.fit_predict(df)

# Separate normal and anomalous points
normal = df[df['anomaly'] == 1]
anomalies = df[df['anomaly'] == -1]

# Plot results
plt.figure()
plt.scatter(normal['distance'], normal['fare'], label='Normal')
plt.scatter(anomalies['distance'], anomalies['fare'], label='Anomaly')
plt.xlabel("Trip Distance")
plt.ylabel("Fare Amount")
plt.title("Isolation Forest ‚Äì Anomaly Detection on NYC Taxi Fare Data")
plt.legend()
plt.show()
```

Question 8: Train a SARIMA model on the monthly airline passengers dataset.
Forecast the next 12 months and visualize the results.
(Include your Python code and output in the code box below.)

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Load AirPassengers dataset
data = pd.read_csv(
    "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv"
)

# Convert Month column to datetime
data['Month'] = pd.to_datetime(data['Month'])
data.set_index('Month', inplace=True)

# Define SARIMA model
# (p,d,q) = (1,1,1)
# (P,D,Q,s) = (1,1,1,12) for yearly seasonality
model = SARIMAX(
    data['Passengers'],
    order=(1, 1, 1),
    seasonal_order=(1, 1, 1, 12)
)

# Fit the model
model_fit = model.fit(disp=False)

# Forecast next 12 months
forecast = model_fit.get_forecast(steps=12)
forecast_mean = forecast.predicted_mean
confidence_intervals = forecast.conf_int()

# Plot original data and forecast
plt.figure()
plt.plot(data['Passengers'], label='Observed')
plt.plot(forecast_mean, label='Forecast')
plt.fill_between(
    confidence_intervals.index,
    confidence_intervals.iloc[:, 0],
    confidence_intervals.iloc[:, 1],
    alpha=0.3
)
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.title("SARIMA Forecast ‚Äì AirPassengers Dataset")
plt.legend()
plt.show()
```

Question 9: Apply Local Outlier Factor (LOF) on any numerical dataset to detect
anomalies and visualize them using matplotlib.
(Include your Python code and output in the code box below.)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor

# Generate a synthetic numerical dataset
np.random.seed(42)

# Normal data points
X_normal = 0.3 * np.random.randn(200, 2)

# Anomalous data points
X_outliers = np.random.uniform(low=-4, high=4, size=(20, 2))

# Combine dataset
X = np.vstack([X_normal, X_outliers])

# Apply Local Outlier Factor
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.1)
y_pred = lof.fit_predict(X)

# Separate normal points and anomalies
normal = X[y_pred == 1]
anomalies = X[y_pred == -1]

# Plot results
plt.figure()
plt.scatter(normal[:, 0], normal[:, 1], label='Normal')
plt.scatter(anomalies[:, 0], anomalies[:, 1], label='Anomaly')
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Local Outlier Factor (LOF) ‚Äì Anomaly Detection")
plt.legend()
plt.show()
```

Question 10: You are working as a data scientist for a power grid monitoring company.
Your goal is to forecast energy demand and also detect abnormal spikes or drops in
real-time consumption data collected every 15 minutes. The dataset includes features
like timestamp, region, weather conditions, and energy usage.
Explain your real-time data science workflow:
‚óè How would you detect anomalies in this streaming data (Isolation Forest / LOF /
DBSCAN)?
‚óè Which time series model would you use for short-term forecasting (ARIMA /
SARIMA / SARIMAX)?
‚óè How would you validate and monitor the performance over time?
‚óè How would this solution help business decisions or operations?
(Include your Python code and output in the code box below.)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
from statsmodels.tsa.statespace.sarimax import SARIMAX

# -------------------------------
# Simulated Energy Consumption Data
# -------------------------------
np.random.seed(42)
date_range = pd.date_range(start="2023-01-01", periods=500, freq="15T")

data = pd.DataFrame({
    "timestamp": date_range,
    "temperature": np.random.normal(25, 5, 500),
    "energy_usage": np.random.normal(300, 30, 500)
})

# Inject anomalies
data.loc[100:105, "energy_usage"] += 200
data.loc[300:305, "energy_usage"] -= 150

data.set_index("timestamp", inplace=True)

# -------------------------------
# Anomaly Detection ‚Äì Isolation Forest
# -------------------------------
iso = IsolationForest(contamination=0.03, random_state=42)
data["anomaly"] = iso.fit_predict(data[["energy_usage", "temperature"]])

anomalies = data[data["anomaly"] == -1]
normal = data[data["anomaly"] == 1]

# Plot anomalies
plt.figure()
plt.plot(data.index, data["energy_usage"], label="Energy Usage")
plt.scatter(anomalies.index, anomalies["energy_usage"], label="Anomaly")
plt.title("Real-Time Energy Anomaly Detection")
plt.xlabel("Time")
plt.ylabel("Energy Usage (kWh)")
plt.legend()
plt.show()

# -------------------------------
# Forecasting ‚Äì SARIMAX
# -------------------------------
model = SARIMAX(
    data["energy_usage"],
    exog=data[["temperature"]],
    order=(1, 1, 1),
    seasonal_order=(1, 1, 1, 96)  # 96 intervals = 1 day (15-min data)
)

model_fit = model.fit(disp=False)

forecast = model_fit.get_forecast(
    steps=96,
    exog=np.random.normal(25, 5, (96, 1))
)

forecast_mean = forecast.predicted_mean

# Plot forecast
plt.figure()
plt.plot(data["energy_usage"], label="Observed")
plt.plot(forecast_mean, label="Forecast")
plt.title("Short-Term Energy Demand Forecast (Next 24 Hours)")
plt.xlabel("Time")
plt.ylabel("Energy Usage (kWh)")
plt.legend()
plt.show()
```
